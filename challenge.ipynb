```python
import pandas as pd
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

#### Criação do DataFrame

```python
df_faturamento = pd.DataFrame().from_dict(dict_faturamento)
```

```python
df_faturamento
```

### Média das vendas

```python
df_faturamento.valor.mean()
```

### Gráfico de barras vertical mostrando o mês de referência e o valor

```python
df_faturamento.info()
```

#### Transformar data_ref para DateTime e formatar para considerar as datas mensalmente

```python
df_faturamento.data_ref = pd.to_datetime(df_faturamento.data_ref).dt.to_period('M')
```

```python
df_faturamento.info()
```

```python
df_faturamento.sort_values(by='data_ref').plot.bar(x='data_ref', y='valor')
```

### Gráfico de linhas mostrando o mês de referência e o valor

```python
df_faturamento.sort_values(by='data_ref').plot.line(x='data_ref', y='valor')
```
